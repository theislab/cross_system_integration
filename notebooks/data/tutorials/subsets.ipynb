```python
import scanpy as sc
import numpy as np
import pandas as pd
import os
```

```python
path_data='/om2/user/khrovati/data/cross_system_integration/'
path_pancreas=path_data+'pancreas_conditions_MIA_HPAP2/'
```

## Randomly subset pancreas adata

```python
adata=sc.read(path_pancreas+'combined_orthologuesHVG.h5ad')
```

```python
pd.crosstab(adata.obs.cell_type_eval,adata.obs.system)
```

```python
n=10000
adata_sub=adata[np.random.RandomState(seed=0).permutation(adata.obs_names)[:n],:]
pd.crosstab(adata_sub.obs.cell_type_eval,adata_sub.obs.system)
```

```python
# Save to tutorials directory
adata_sub.write(os.getcwd().split('notebooks/')[0]+'tutorials/data/'+f'mouse-human_pancreas_subset{n}.h5ad')
```
